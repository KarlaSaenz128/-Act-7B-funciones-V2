```python
# Problema 48: Cálculo de FCM y Esfuerzo Físico
# Usando if, elif, while y validaciones profesionales

def calcular_transferencia_calor():
    while True:
        print("\n--- CALCULADORA DE TRANSFERENCIA DE CALOR ---")
        print("1. Calcular Calor Absorbido (Q)")
        print("2. Calcular Variación de Temperatura (ΔT)")
        print("3. Salir")

        opcion = input("\nSeleccione una opción: ")

        if opcion == "3":
            print("Programa finalizado. ¡Cuidado con las altas temperaturas!")
            break

        if opcion == "1" or opcion == "2":
            try:
                if opcion == "1":
                    # Cálculo de Q = m * Ce * ΔT
                    print("\n>>> CÁLCULO DE CALOR ABSORBIDO (Q) <<<")
                    m = float(input("Ingrese la masa (m): "))
                    ce = float(input("Ingrese el calor específico (Ce): "))
                    dt = float(input("Ingrese la variación de temperatura (ΔT): "))

                    q = m * ce * dt
                    print(f"\nRESULTADO: El calor absorbido es: {q:.2f} unidades.")

                else:
                    # Cálculo de ΔT = Q / (m * Ce)
                    print("\n>>> CÁLCULO DE VARIACIÓN DE TEMPERATURA (ΔT) <<<")
                    q = float(input("Ingrese el calor (Q): "))
                    m = float(input("Ingrese la masa (m): "))
                    ce = float(input("Ingrese el calor específico (Ce): "))

                    if m * ce == 0:
                        print("Error: La masa o el calor específico no pueden ser cero.")
                    else:
                        dt = q / (m * ce)
                        print(f"\nRESULTADO: La variación de temperatura es: {dt:.2f} °C.")

            except ValueError:
                print("Error: Por favor, ingrese valores numéricos válidos.")
        else:
            print("Opción no válida. Intente de nuevo.")
            print("Estructura original basada en el programa de Karla Saenz 0133.")

def calcular_fcm_y_esfuerzo():
    while True:
        # Inicializamos las variables en None para cada iteración (limpiar pizarra)
        fcm = None
        frecuencia_actual = None

        print("\n" + "*"*10 + " MENÚ DE SALUD " + "*"*10)
        print("1. Fórmula para hombres (220 - edad)")
        print("2. Fórmula para mujeres (226 - edad)")
        print("3. Salir")
        print("*-*-*-*-*")

        opcion = input("Ingrese una opcion: ") # Usamos input para evitar errores si no es número

        if opcion == "1":
            print("\n--- Modo: Hombre ---")
            try:
                edad = int(input("Ingrese su edad: "))
                fcm = 220 - edad
                frecuencia_actual = int(input("Ingrese su frecuencia cardíaca actual (lpm): "))
            except ValueError:
                print("Error: Por favor, ingrese valores numéricos válidos para edad y frecuencia cardíaca.")
                continue

        elif opcion == "2":
            print("\n--- Modo: Mujer ---")
            try:
                edad = int(input("Ingrese su edad: "))
                fcm = 226 - edad
                frecuencia_actual = int(input("Ingrese su frecuencia cardíaca actual (lpm): "))
            except ValueError:
                print("Error: Por favor, ingrese valores numéricos válidos para edad y frecuencia cardíaca.")
                continue

        elif opcion == "3":
            print("Saliendo del programa... ¡Cuida tu corazón!")
            break

        else:
            print("❌ Opción no válida. Intente de nuevo.")
            continue # Reinicia el bucle para pedir una opción válida

        # Usando porcentaje_esfuerzo por consistencia y evitando cálculos duplicados
        # Solo procedemos si fcm y frecuencia_actual tienen valores válidos (no son None)
        if fcm is not None and frecuencia_actual is not None:
            if fcm == 0:
                print("Error: FCM no puede ser cero para calcular el porcentaje de esfuerzo.")
            else:
                porcentaje_esfuerzo = (frecuencia_actual / fcm) * 100

                print("-" * 30)
                print(f"Tu FCM teórica es: {fcm} lpm")
                print(f"Tu nivel de esfuerzo actual es del: {porcentaje_esfuerzo:.2f}%")

                # Clasificación rápida del esfuerzo
                if porcentaje_esfuerzo < 60:
                    print("Intensidad: Baja (Calentamiento)")
                elif 60 <= porcentaje_esfuerzo < 85:
                    print("Intensidad: Media (Mejora de salud)")
                else:
                    print("Intensidad: Alta (Zona máxima)")
                print("-" * 30)

        print("Gracias por usar el programa")
        print("programa realizado por Saenz Karla")

# Llamada a las funciones
if __name__ == "__main__":
    calcular_transferencia_calor()
    calcular_fcm_y_esfuerzo()
```
