### Calculadora de Sueño
#### Calcula a qué hora despertar si te duermes ahora (ciclos de 90 min)
1. Después de 4 ciclos
2. Después de 5 ciclos
3. Después de 6 ciclos
4. Salir.


```python
# Problema 51: Calculadora de Ciclos de Sueño
# Karla Saenz 0133
# Estructura profesional con validaciones y manejo de tiempo
from datetime import datetime, timedelta

def calcular_ciclos_sueno():
    while True:
        # Limpiamos la pantalla visualmente con el menú
        print("\n" + "🌙" * 5 + " CALCULADORA DE SUEÑO REPARADOR " + "🌙" * 5)
        print("Si te duermes ahora, ¿a qué hora deberías despertar?")
        print("1. Después de 4 ciclos (6 horas - Descanso mínimo)")
        print("2. Después de 5 ciclos (7.5 horas - Descanso ideal)")
        print("3. Después de 6 ciclos (9 horas - Descanso completo)")
        print("4. Salir")
        print("-" * 40)

        opcion = input("Seleccione una opción: ")

        if opcion == "4":
            print("\n¡Buenas noches! Recuerda apagar las pantallas antes de dormir.")
            break

        # Validamos que la opción sea correcta antes de calcular
        if opcion in ["1", "2", "3"]:
            # Obtenemos la hora actual
            ahora = datetime.now()

            # Definimos cuántos ciclos sumar según la opción
            if opcion == "1":
                num_ciclos = 4
                descripcion = "4 ciclos (Descanso corto)"
            elif opcion == "2":
                num_ciclos = 5
                descripcion = "5 ciclos (Descanso recomendado)"
            else:
                num_ciclos = 6
                descripcion = "6 ciclos (Descanso profundo)"

            # Cálculo: Cada ciclo dura 90 minutos
            minutos_totales = num_ciclos * 90

            # Se agregan 15 minutos adicionales que es el tiempo promedio en quedarse dormido
            hora_despertar = ahora + timedelta(minutes=minutos_totales + 15)

            # Formateo de resultados
            print("\n" + "=" * 35)
            print(f"ANÁLISIS PARA: {descripcion}")
            print(f"Hora actual: {ahora.strftime('%H:%M')}")
            print(f"Tiempo total de sueño: {minutos_totales / 60} horas")
            print(f">>> DEBERÍAS DESPERTAR A LAS: {hora_despertar.strftime('%H:%M')} <<<")
            print("=" * 35)
            print("Nota: Se han incluido 15 min. para conciliar el sueño.")

        else:
            print("\n❌ Opción no válida. Por favor, elige del 1 al 4.")

        # Pie de autoría solicitado
        print("\n" + "." * 20)
        print("Programa elaborado por Karla Saenz")
        print("Lenguaje: Python | Nivel: Preparatoria")
        print("." * 20)

# Ejecución del programa
if __name__ == "__main__":
    calcular_ciclos_sueno()
```
