```python
# Algoritmo para obtener el promedio de 4 calificaciones

def VerMenu():
    print("--- MENÚ DE OPCIONES ---")
    print("1. Calcular Promedio de 4 calificaciones")
    print("2. Salir")
    opcion = int(input("Ingrese la opcion deseada: "))
    return opcion

def PromedioCalificaciones():
    print("--- Ejemplo 2.2: Cálculo de Promedio ---")

    # 1. Leer C1, C2, C3, C4 (Tipo Real/float)
    C1 = float(input("Ingrese la calificación 1 (C1): "))
    C2 = float(input("Ingrese la calificación 2 (C2): "))
    C3 = float(input("Ingrese la calificación 3 (C3): "))
    C4 = float(input("Ingrese la calificación 4 (C4): "))

    # 2. Hacer S = C1 + C2 + C3 + C4
    S = C1 + C2 + C3 + C4

    # 3. Hacer P = S / 4
    P = S / 4

    # 4. Escribir P
    print("Procesando datos...")
    print("La suma total (S) es:", S)
    print("El promedio final (P) es:", P)

# --- Ciclo principal ---
while True:
    OP = VerMenu()
    if OP == 1:
        PromedioCalificaciones()
    elif OP == 2:
        print("Gracias por usar el programa")
        break
    else:
        print("Opción no válida, intente de nuevo.")

print("Programa realizado por karla saenz 0133")
print("Utilizando la inteligencia Artificial Gemini de Google")

# Menu del problema: Área de una circunferencia

def VerMenu():
  print("--- MENÚ DE OPCIONES ---")
  print("1. Calcular Área de una circunferencia (Ejemplo 2.4)")
  print("2. Salir")
  opcion = int(input("Ingrese la opcion deseada: "))
  return opcion

def AreaCircunferencia():
  print("--- Ejecutando Algoritmo 2.4 ---")

  # 1. Leer R (Radio de la circunferencia - Tipo Real)
  R = float(input("Ingrese el valor del radio (R): "))

  # 2. Hacer PI = 3.1416
  PI = 3.1416

  # 3. Hacer Área = PI * R * R (Fórmula: PI * R2)
  Area = PI * R * R

  # 4. Escribir Área
  print("Calculando el área...")
  print("El radio ingresado es:", R)
  print("El área calculada es:", Area)

# --- Ciclo principal del programa ---
while True:
  OP = VerMenu()
  if OP == 1:
    AreaCircunferencia()
  elif OP == 2:
    print("Gracias por usar el programa")
    break
  else:
    print("Opción no válida, intente de nuevo")

print("Programa realizado por karla saenz 0133")
print("Utilizando la inteligencia Artificial Gemini de Google")

import math # Necesario para usar math.sqrt (Raíz cuadrada)

# Menu del problema: Área de la figura compuesta (Ejemplo 2.6)

def VerMenu():
  print("--- MENÚ DE OPCIONES ---")
  print("1. Calcular Área de figura compuesta (Ejemplo 2.6)")
  print("2. Salir")
  opcion = int(input("Ingrese la opcion deseada: "))
  return opcion

def CalcularAreaCompuesta():
  print("--- Ejecutando Algoritmo 2.6 ---")

  # 1. Leer R, H (Radio/Base e Hipotenusa)
  R = float(input("Ingrese el valor del Radio/Base (R): "))
  H = float(input("Ingrese el valor de la Hipotenusa (H): "))

  # Validación matemática: H debe ser mayor que R para la raíz cuadrada
  if H > R:
    # 2. Hacer C = SQRT (H * H - R * R) -> Cateto faltante
    C = math.sqrt(H**2 - R**2)

    # 3. Hacer AT = 2 * (R * C) / 2 -> Área de los dos triángulos
    AT = 2 * (R * C) / 2

    # 4. Hacer AC = (3.1416 * R * R) / 2 -> Área del semicírculo
    PI = 3.1416
    AC = (PI * R * R) / 2

    # 5. Hacer Área = AT + AC
    AreaTotal = AT + AC

    # 6. Escribir Área
    print("--- Resultados ---")
    print(f"Cateto calculado (C): {C:.2f}")
    print(f"Área Triangular (AT): {AT:.2f}")
    print(f"Área Circular (AC): {AC:.2f}")
    print(f"El Área Total de la figura es: {AreaTotal:.2f}")
  else:
    print("Error: La hipotenusa (H) debe ser mayor que el radio (R).")

# --- Ciclo principal ---
while True:
  OP = VerMenu()
  if OP == 1:
    CalcularAreaCompuesta()
  elif OP == 2:
    print("Gracias por usar el programa")
    break
  else:
    print("Opción no válida")

print("Programa realizado por karla saenz 0133")
print("Utilizando la inteligencia Artificial Gemini de Google")
import math # Necesario para SQRT (raíz cuadrada)

# Menu del problema: Distancia entre dos puntos (Ejemplo 2.8)

def VerMenu():
  print("--- MENÚ DE OPCIONES ---")
  print("1. Calcular Distancia entre 2 puntos (Ejemplo 2.8)")
  print("2. Salir")
  opcion = int(input("Ingrese la opcion deseada: "))
  return opcion

def CalcularDistancia():
  print("--- Ejecutando Algoritmo 2.8 ---")

  # 1. Leer coordenadas del punto 1 (X1, Y1)
  X1 = float(input("Ingrese coordenada X1: "))
  Y1 = float(input("Ingrese coordenada Y1: "))

  # 2. Leer coordenadas del punto 2 (X2, Y2)
  X2 = float(input("Ingrese coordenada X2: "))
  Y2 = float(input("Ingrese coordenada Y2: "))

  # 3. Hacer X = X2 - X1 (Cateto de las abscisas)
  X = X2 - X1

  # 4. Hacer Y = Y2 - Y1 (Cateto de las ordenadas)
  Y = Y2 - Y1

  # 5. Hacer D = SQRT (X * X + Y * Y)
  # Usamos math.sqrt para la raíz cuadrada
  D = math.sqrt(X*X + Y*Y)

  # 6. Escribir D
  print("--- Resultados ---")
  print(f"Diferencia en X: {X}")
  print(f"Diferencia en Y: {Y}")
  print(f"La distancia entre los puntos (D) es: {D:.2f}")

# --- Ciclo principal ---
while True:
  OP = VerMenu()
  if OP == 1:
    CalcularDistancia()
  elif OP == 2:
    print("Gracias por usar el programa")
    break
  else:
    print("Opción no válida, intente de nuevo.")

print("Programa realizado por karla saenz 0133")
print("Utilizando la inteligencia Artificial Gemini de Google")

# Menu del problema: Conversión de Metros a Pulgadas

def VerMenu():
  print("--- MENÚ DE OPCIONES ---")
  print("1. Convertir Metros a Pulgadas (Ejemplo 2.10)")
  print("2. Salir")
  opcion = int(input("Ingrese la opcion deseada: "))
  return opcion

def ConvertirMetrosPulgadas():
  print("--- Ejecutando Algoritmo 2.10 ---")

  # 1. Leer CM (Cantidad de metros - Tipo Real)
  CM = float(input("Ingrese la cantidad de metros (CM): "))

  # 2. Hacer PG = CM / 0.0254
  # Nota: Se usa 0.0254 porque 1 pulgada mide 0.0254 metros
  PG = CM / 0.0254

  # 3. Escribir PG
  print("--- Resultado de la conversión ---")
  print(f"Metros ingresados: {CM} m")
  print(f"Pulgadas a pedir (PG): {PG:.4f} in")

# --- Ciclo principal del programa ---
while True:
  OP = VerMenu()
  if OP == 1:
    ConvertirMetrosPulgadas()
  elif OP == 2:
    print("Gracias por usar el programa")
    break
  else:
    print("Opción no válida, intente de nuevo")

print("Programa realizado por karla saenz 0133")
print("Utilizando la inteligencia Artificial Gemini de Google")
```
